```python
import matplotlib.pyplot as plt
import numpy as np



datetime, internal, external, tds, sal, conduc, batt = np.genfromtxt('C:\\Users\\noah_\\Downloads\\PYTHON\\Senior Design Graph Code\\DATA 04122020.csv', dtype = str, delimiter =",", unpack = True)

datetime = np.delete(datetime,0)
internal = np.delete(internal,0).astype(np.float)
external = np.delete(external,0).astype(np.float)
tds      = np.delete(tds,0).astype(np.float)
sal      = np.delete(sal,0).astype(np.float)
conduc   = np.delete(conduc,0).astype(np.float)
batt     = np.delete(batt,0).astype(np.float)



x_ticks  = np.arange(1,len(datetime), len(datetime)/30)
batt_ticks = np.arange(1,len(datetime), len(datetime)/20)
batt_ticks2 = np.arange(1,len(datetime), len(datetime)/30)
x_axis = []

avg_voltage = 0
total = 0

temp_high = np.amax(internal)
#print(external[0],f"Max Temp was {temp_high}")

#np.where(internal[i]==0, internal[i-1], internal)
#print(internal[514])


# uncommend the following code to erease the error values of -1. THe graph
# will take the last valid reading and push it through until the error has gone away.

for i in range(len(datetime)):
    
    if tds[i] == -1:
        tds[tds < 0] = tds[i-1]
    else:
        pass
    
    if external[i] == -1:
        external[external < 0] = external[i-1]
    else:
        pass   
    
    if internal[i] == -1:
        internal[internal < 0] = internal[i-1]
    else:
        pass
    
    if sal[i] == -1:
        sal[sal < 0] = sal[i-1]
    else:
        pass
    
for i in range(len(batt[902:-1:1])):
    total = total+batt[902+i]
    avg_batt = round(total/len(batt[902:-1]),2)

plt.figure(figsize=(20,10)) #total size of the graph
plt.plot(datetime, internal, color = 'purple', label = 'Internal Temperature',linestyle='-',markerfacecolor='k', linewidth = 1)
plt.plot(datetime,external, color = 'red', label = 'External Temperature',linestyle='-',markerfacecolor='k', linewidth = 1)
plt.xlabel("Date and Time", fontsize = 30)
plt.ylabel("Temperature (°f)",fontsize = 30)
plt.title("Temperature Over Time", family='serif', c='black',fontsize = 50, fontweight='bold')
plt.xticks(x_ticks, ha = "left",rotation=-45, fontsize = 15) #rotating the x axis label
plt.yticks(rotation=90, fontsize = 15) #rotating the y axis label
plt.gcf().subplots_adjust(bottom=0.20)
plt.legend()
plt.savefig('plotTemps4Samples.jpg')
plt.show()


plt.figure(figsize=(20,10)) #total size of the graph
plt.plot(datetime,tds, color = 'purple',linestyle='-',markerfacecolor='k', linewidth = 1)
plt.xlabel("Date and Time", fontsize = 30)
plt.ylabel("Total Dissolved Solids,(ppm)", fontsize = 30)
plt.title("Total Dissolved Solids in the Water", family='serif', c='black',fontsize = 50, fontweight='bold')
plt.xticks(x_ticks,ha = "left",rotation=-45) #rotating the x axis label
plt.yticks(rotation=90) #rotating the y axis label
plt.legend()
plt.savefig('plotTDS4Samples.jpg')
plt.show()

plt.figure(figsize=(20,10)) #total size of the graph
plt.plot(datetime,sal, color = 'red',linestyle='-',markerfacecolor='k', linewidth = 1)
plt.xlabel("Date and Time", fontsize = 30)
plt.ylabel("Salinity,(psu g/kg)", fontsize = 30)
plt.title("Salinity of the Water", family='serif', c='black',fontsize = 50, fontweight='bold')
plt.xticks(x_ticks,ha = "left",rotation=-45) #rotating the x axis label
plt.yticks(rotation=90) #rotating the y axis label
plt.legend()
plt.savefig('plotSalinity4Samples.jpg')
plt.show()

plt.figure(figsize=(20,10)) #total size of the graph
plt.plot(datetime,conduc, color = 'red',linestyle='-',markerfacecolor='k', linewidth = 1)
plt.xlabel("Date and Time",fontsize = 30)
plt.ylabel("Conductivity (μS/cm)",fontsize = 30)
plt.title("Conductivity of the Water", family='serif', c='black',fontsize = 50, fontweight='bold')
plt.xticks(x_ticks,ha = "left",rotation=-45) #rotating the x axis label
plt.yticks(rotation=90) #rotating the y axis label
plt.legend()
plt.savefig('plotConductivity4Samples.jpg')
plt.show()


plt.figure(figsize=(20,10)) #total size of the graph
plt.plot(datetime,batt, color = 'red',linestyle = '-',markerfacecolor='k', linewidth = 1)
plt.xlabel("Date and Time", fontsize = 30)
plt.ylabel("Battery Voltage(V)", fontsize = 30)
plt.title("Battery Voltage", family='serif', c='black',fontsize = 50, fontweight='bold')
plt.xticks(batt_ticks,ha = "left",rotation=-45) #rotating the x axis label
plt.yticks(rotation=90) #rotating the y axis label
plt.ylim(3.9,4.5)
plt.xlim(0,datetime[-1])
#plt.text(datetime[915],batt[900], "Voltage reading fixed.")
plt.annotate("Voltage reading repaired.",(datetime[915],4.14),(datetime[930],batt[900]),arrowprops=dict(arrowstyle="->"))
plt.savefig('plotBatteryVoltagefull4Samples.jpg')
plt.show()




plt.figure(figsize=(20,10)) #total size of the graph
plt.plot(datetime,batt, color = 'red',linestyle = '-',markerfacecolor='k', linewidth = 1)
plt.xlabel("Date and Time", fontsize = 30)
plt.ylabel("Battery Voltage(V)", fontsize = 30)
plt.title("Battery Voltage, Two Week Period", family='serif', c='black',fontsize = 50, fontweight='bold')
plt.xticks(batt_ticks,ha = "left",rotation=-45) #rotating the x axis label
plt.yticks(rotation=90) #rotating the y axis label
plt.ylim(4.04,4.205)
plt.xlim(datetime[902],datetime[-1])
plt.text(datetime[950],4.145,f"Average battery voltage of {avg_batt}")
plt.annotate(f" ", (datetime[902],avg_batt),(datetime[-1],avg_batt),arrowprops=dict(arrowstyle="<->"))
plt.savefig('plotBatteryVoltageWorking4Samples.jpg')
plt.show()

#Average battery voltage of: \n {avg_batt}


```
